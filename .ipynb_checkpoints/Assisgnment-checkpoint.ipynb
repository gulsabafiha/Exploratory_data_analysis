```python
import matplotlib.pyplot as plt

# Assuming the DataFrame is named "df"
mileage = car_data['mileage']

plt.hist(mileage, bins=10, edgecolor='black')
plt.xlabel('Mileage')
plt.ylabel('Frequency')
plt.title('Distribution of Mileage')
plt.show()
```

```python
# Assuming the DataFrame is named "df"
car_data['mileage'].plot(kind='bar')

plt.xlabel('Index')
plt.ylabel('Mileage')
plt.title('Mileage Bar Plot')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the DataFrame is named "df"
plt.figure(figsize=(10, 4))  # Set the figure size

# Scatter plot for Price vs. Mileage
plt.subplot(1, 2, 1)  # Create the first subplot
plt.scatter(car_data['price'], car_data['mileage'])
plt.xlabel('Price')
plt.ylabel('Mileage')
plt.title('Price vs. Mileage')

# Scatter plot for Mileage vs. Price
plt.subplot(1, 2, 2)  # Create the second subplot
plt.scatter(car_data['mileage'], car_data['price'])
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('Mileage vs. Price')

plt.tight_layout()  # Adjust the spacing between subplots

plt.show()
```

```python

import seaborn as sns

# Assuming the DataFrame is named "df"
sns.regplot(x='mileage', y='price', data=car_data)
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('Regression Plot: Price vs Mileage')

plt.figure()  # Create a new figure for the second plot

sns.regplot(x='price', y='mileage', data=car_data)
plt.xlabel('Price')
plt.ylabel('Mileage')
plt.title('Regression Plot: Mileage vs Price')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


## Create a Regression plot using two numeric columns

data = pd.read_csv('bmw.csv')
df = pd.DataFrame(data)
print(df.columns)

d = np.polyfit(df['year'],df['price'],1)
f = np.poly1d(d)
print(d)
print(f)

ax = df.plot(x = 'year',y='price')
df.plot(x='year', y='price',color='purple',ax=ax)
```
