```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('bmw.csv')
# print(df.columns)



plt.scatter(df['price'], df['year'], s=100, alpha=0.6, edgecolor='black', linewidth=1)

plt.title('Scatter Plot for Price regarding Year')
plt.xlabel('Price')
plt.ylabel('Year')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


## Create a Regression plot using two numeric columns

data = pd.read_csv('bmw.csv')
df = pd.DataFrame(data)
print(df.columns)

d = np.polyfit(df['year'],df['price'],1)
f = np.poly1d(d)
print(d)
print(f)

ax = df.plot(x = 'year',y='price')
df.plot(x='year', y='price',color='purple',ax=ax)
```

```python
# 4. Create a Pair plot with numeric columns. 

sns.pairplot(df, hue = 'fuelType')

### This is a pairplot with numeric columns
```
